```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn import svm
```

```python
#Loading the dataset
data=pd.read_csv('Parkinsons.csv')
```

```python
#viewing the data
data.head()
```

```python
#knowing the shape of data
data.shape
```

```python
#knowing the info of the data
data.info()
```

```python
#describing the data
data.describe()
```

```python
#counting null values for each column
data.isnull().sum()
```

```python
#counting the targer values
data['status'].value_counts()
```

```python
#viewing the target values and their means
data.select_dtypes(include=np.number).groupby('status').mean()
```

```python
#data preprocessing
#seperating features and target
x= data.drop(columns=['name','status'],axis=1)  #dropping name column as it is not required
y= data['status']
x.head(),y.head()
```

```python
#splitting data in training and testing data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
x.shape,x_train.shape,x_test.shape
```

```python
#Data standardization
scaler=StandardScaler() 
scaler.fit(x_train)
```

```python
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
#model training
#support vector machine model
model=svm.SVC(kernel='linear')
```

```python
#training the svm model
model.fit(x_train,y_train)
```

```python
#Accuracy of training data
x_train_prediction=model.predict(x_train)
x_train_predict_accuracy=accuracy_score(x_train_prediction,y_train)
x_train_predict_accuracy
```

```python
#Accuracy of traiing data
x_test_prediction=model.predict(x_test)
x_test_predict_accuracy=accuracy_score(x_test_prediction,y_test)
x_test_predict_accuracy
```

```python
#building a predictive system
input_data=(116.014,141.781,110.655,0.01284,0.00011,0.00655,0.00908,0.01966,0.06425,0.584,0.0349,0.04825,0.04465,0.1047,0.01767,19.649,0.417356,0.823484,-3.747787,0.234513,2.33218,0.410335)
#changing input data to a numpy array
input_data_np=np.asarray(input_data)
# reshaping numpy array
input_data_reshaped=input_data_np.reshape(1,-1) #giving the data of one dtaa point otherwise it except 196  datas as above
# standardize the data
std_data=scaler.transform(input_data_reshaped)
# prediction
prediction=model.predict(std_data)
print(prediction)

if prediction[0]==0:
  print('The person does not have Parkinsons Disease')
else:
  print('The person has Parkinsons Disease')
# input_data
```
