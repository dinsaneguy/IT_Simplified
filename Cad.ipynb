```python


# 1=Yes 0=No
data['Obesity'] = data['Obesity'].map({'Y': 1, 'N': 0})
data['CRF'] = data['CRF'].map({'Y': 1, 'N': 0})
data['CVA'] = data['CVA'].map({'Y': 1, 'N': 0})
data['Airway disease'] = data['Airway disease'].map({'Y': 1, 'N': 0})
data['Thyroid Disease'] = data['Thyroid Disease'].map({'Y': 1, 'N': 0})
data['CHF'] = data['CHF'].map({'Y': 1, 'N': 0})
data['DLP'] = data['DLP'].map({'Y': 1, 'N': 0})
# data[''] = data[''].map({'Y': 1, 'N': 0})
data['Weak Peripheral Pulse'] = data['Weak Peripheral Pulse'].map({'Y': 1, 'N': 0})
data['Lung rales'] = data['Lung rales'].map({'Y': 1, 'N': 0})
data['Systolic Murmur'] = data['Systolic Murmur'].map({'Y': 1, 'N': 0})
data['Diastolic Murmur'] = data['Diastolic Murmur'].map({'Y': 1, 'N': 0})
data['Dyspnea'] = data['Dyspnea'].map({'Y': 1, 'N': 0})
data['Atypical'] = data['Atypical'].map({'Y': 1, 'N': 0})
data['Nonanginal'] = data['Nonanginal'].map({'Y': 1, 'N': 0})
data['Exertional CP'] = data['Exertional CP'].map({'Y': 1, 'N': 0})
data['LowTH Ang'] = data['LowTH Ang'].map({'Y': 1, 'N': 0})
data['LVH'] = data['LVH'].map({'Y': 1, 'N': 0})
data['Poor R Progression'] = data['Poor R Progression'].map({'Y': 1, 'N': 0})
data['Age']=data['Age'].astype('float64')
data['Sex'] = data['Sex'].map({'Male': 1, 'Fmale': 0})  #1=Male , 0=Female


data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
data['VHD'] = label_encoder.fit_transform(data['VHD'])
data['VHD'].isna().sum()
print(data['VHD'].value_counts())

#0=N, 1=mild, 2=severe, 3=moderate
```

```python
x=data.drop('Cath',axis=1)
y=data['Cath']
x.isnull().sum()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
x_train.shape,x_test.shape,y_train.shape,y_test.shape,x.shape,y.shape
```

```python

x_train.head()
```

```python
scalar=StandardScaler()
x_train=scalar.fit_transform(x_train)
```

```python
x_train=scalar.transform(x_train)
x_test=scalar.transform(x_test)
x_train,x_test
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
x_train_prediction=model.predict(x_train)
x_train_accuracy=accuracy_score(x_train_prediction,y_train)
x_train_accuracy
```

```python
x_test_prediction=model.predict(x_test)
x_test_accuracy=accuracy_score(x_test_prediction,y_test)
x_test_accuracy
```

```python
#changint outcome as 1=Cad ans 0=Normal
data['Cath'] = data['Cath'].map({'Cad': 1, 'Normal': 0})
```
